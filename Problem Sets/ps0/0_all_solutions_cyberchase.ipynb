## 01  
Write a SQL query to list the titles of all episodes in Cyberchase’s original season, Season 1.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "season" IS 1;
```

## 02  
List the season number of, and title of, the first episode of every season.

```python
%%sql
SELECT "season", "title"
FROM "episodes"
WHERE "episode_in_season" == 1;
```

## 03  
Find the production code for the episode “Hackerized!”.

```python
%%sql
SELECT "production_code"
FROM "episodes"
WHERE "title" == 'Hackerized!';
```

## 04  
Write a query to find the titles of episodes that do not yet have a listed topic.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "topic" IS NULL;
```

## 05  
Find the title of the holiday episode that aired on December 31st, 2004.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "air_date" IS '2004-12-31';
```

## 06  
List the titles of episodes from season 6 (2008) that were released early, in 2007.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "air_date" LIKE '2007-%'
AND "season" IS 6;
```

## 07  
Write a SQL query to list the titles and topics of all episodes teaching fractions.

```python
%%sql
SELECT "title", "topic"
FROM "episodes"
WHERE "topic" LIKE '%fractions%';
```

## 08  
Write a query that counts the number of episodes released in the last 6 years, from 2018 to 2023, inclusive.

```python
%%sql
SELECT COUNT("title")
FROM "episodes"
WHERE "air_date"
BETWEEN '2018-01-01' AND '2023-12-31';
```

## 09

## 10  
Write a SQL query to list the ids, titles, and production codes of all episodes. Order the results by production code, from earliest to latest.

```python
%%sql
SELECT "id", "title", "production_code"
FROM "episodes"
ORDER BY "production_code" ASC;
```

## 11  
List the titles of episodes from season 5, in reverse alphabetical order.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "season" IS '5'
ORDER BY "title" DESC;
```

## 12

Write a SQL query to explore a question of your choice. This query should:

Question: Write a SQL query to retrieve the titles and topics of episodes published between 2002 and 2010 (inclusive) that are on the topic 'data'.

```python
%%sql
SELECT "title"
FROM "episodes"
WHERE "air_date"
BETWEEN '2002-01-01' AND '2010-12-31'
AND "topic" LIKE '%data%';
```

Write a SQL query to find, for each year, the first day of the year that PBS released a Cyberchase episode.
- Your query should output a table with two columns, one for the year and one for the earliest month and day an episode was released that year.

```python
%%sql
SELECT
  strftime('%Y', "air_date") AS 'year',
  MIN(strftime('%m-%d', "air_date")) AS 'earliest_release'
FROM "episodes"
GROUP BY strftime('%Y', "air_date");
```
