## Joins

```python
database_name = "sea_lions"
```

```python
conn = sqlite3.connect(f"./{database_name}.db")
conn.close()
```

```python
%sql sqlite:///{database_name}.db?check_same_thread=false
%sql SELECT name FROM sqlite_master WHERE type='table';
```

```python
%%sql
SELECT * FROM "sea_lions";
```

```python
%%sql
SELECT * FROM "migrations";
```

### INNER JOIN

```python
%%sql
SELECT * FROM "sea_lions"
JOIN "migrations" 
ON "migrations"."id" = "sea_lions"."id";
```

### LEFT JOIN

```python
%%sql
SELECT * FROM "sea_lions"
LEFT JOIN "migrations" 
ON "migrations"."id" = "sea_lions"."id";
```

### RIGHT JOIN

```python
%%sql
SELECT * FROM "sea_lions"
RIGHT JOIN "migrations" 
ON "migrations"."id" = "sea_lions"."id";
```

### FULL JOIN

```python
%%sql
SELECT * FROM "sea_lions"
FULL JOIN "migrations" 
ON "migrations"."id" = "sea_lions"."id";
```

### NATURAL JOIN

```python
%%sql
SELECT * FROM "sea_lions"
NATURAL JOIN "migrations";
```

### UNION

## Groups

```python
%%sql
SELECT AVG(rating) FROM "ratings";
```

```python
%%sql
SELECT "book_id", AVG("rating") AS "average_rating" FROM "ratings" 
GROUP BY "book_id";
```

### HAVING

```python
%%sql
SELECT "book_id", ROUND(AVG("rating"), 2) AS "average_rating"
FROM "ratings"
GROUP BY "book_id"
HAVING "average_rating" >= 4.0;
```

```python
%%sql
SELECT "book_id", COUNT("rating") AS "num_ratings" FROM "ratings" 
GROUP BY "book_id";
```

```python
%%sql
SELECT "book_id", ROUND(AVG("rating"), 2) AS "average_rating"
FROM "ratings"
GROUP BY "book_id"
HAVING "average_rating" >= 4.0
ORDER BY "average_rating" DESC;
```
