## Aggregation Function
`COUNT, AVG, MIN, MAX, SUM`

```python
%%sql
SELECT AVG("rating")
FROM "longlist";
```

```python
%%sql
SELECT ROUND(AVG("rating"), 2) AS "average rating"
FROM "longlist";
```

```python
%%sql
SELECT MAX("rating")
FROM "longlist";
```

```python
%%sql
SELECT MIN("rating")
FROM "longlist";
```

```python
%%sql
SELECT SUM("votes")
FROM "longlist";
```

```python
%%sql
SELECT COUNT(*)
FROM "longlist";
```

```python
%%sql
SELECT COUNT("translator")
FROM "longlist";
```

```python
%%sql
SELECT COUNT("publisher")
FROM "longlist";
```

```python
%%sql
SELECT DISTINCT "publisher"
FROM "longlist";
```

```python
%%sql
SELECT COUNT(DISTINCT "publisher")
FROM "longlist";
```
